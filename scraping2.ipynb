```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import requests 
import time
```

```python
driver = webdriver.Chrome()

nama_produk = []
harga_produk = []
penjual_toko = []
kota_toko = []
banyaknya_terjual = []
list_rating = []
for page in range(1,21):
    web_link = f'https://www.tokopedia.com/search?st=&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&navsource={page}'
    driver.get(web_link)
    time.sleep(10)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    petak_list = soup.find_all('div', {'class':"css-5wh65g"})
    for petak in petak_list:
        try:
            nama_produk.append(petak.find('div', {'class':'bYD8FcVCFyOBiVyITwDj1Q=='}).get_text())
        except:
            nama_produk.append(None)
        try:
            harga_produk.append(petak.find('div', {'class':'XvaCkHiisn2EZFq0THwVug=='}).get_text())
        except:
            harga_produk.append(None)
        try:
            penjual_toko.append(petak.find('span', {'class':'T0rpy-LEwYNQifsgB-3SQw== pC8DMVkBZGW7-egObcWMFQ== flip'}).get_text())
        except:
            penjual_toko.append(None)
        try:
            kota_toko.append(petak.find('span', {'class':'pC8DMVkBZGW7-egObcWMFQ== flip'}).get_text())
        except:
            kota_toko.append(None)
        try:
            banyaknya_terjual.append(petak.find('span', {'class':'se8WAnkjbVXZNA8mT+Veuw=='}).get_text())
        except:
            banyaknya_terjual.append(None)
        try:
            list_rating.append(petak.find('div', {'class':'_7wVAVu6CWnC3tsW+zT0z5g=='}).get_text())
        except:
            list_rating.append(None)

driver.close()
```

```python
df = pd.DataFrame() 
```

```python
df['Nama Produk'] = nama_produk
df['Harga Produk'] = harga_produk
df['Penjual'] = penjual_toko
df['Kota Toko'] = kota_toko
df['Banyaknya Terjual'] = banyaknya_terjual
df['Rating Produk'] = list_rating
```

```python
df.head(50)
```

```python
df['Harga Produk'] = df['Harga Produk'].str.extract(r'(Rp\d+\.\d+|Rp\d+)$')
```

```python
# meubah data pada kolom harga supaya tidak terdapat karakter lain selain angka
df['Harga Produk'] = df['Harga Produk'].str.replace("Rp", "")
df['Harga Produk'] = df['Harga Produk'].str.replace(".", "")
```

```python
# Ekstrak angka pertama dari string (termasuk desimal)
df['Banyaknya Terjual'] = df['Banyaknya Terjual'].str.extract(r'(\d+\.?\d*)')[0]

# Isi NaN dengan 0 agar tidak error saat dikonversi
df['Banyaknya Terjual'] = df['Banyaknya Terjual'].fillna(0).astype(float)

# Kalikan 1000 jika ada 'rb' dalam teks
df['Banyaknya Terjual'] = df['Banyaknya Terjual'] * df['Banyaknya Terjual'].astype(str).str.contains('rb').replace({True: 1000, False: 1})

# Ubah ke integer
df['Banyaknya Terjual'] = df['Banyaknya Terjual'].astype(int)
```

```python
# Hapus baris yang memiliki NaN di semua kolom
df = df.dropna(how='all')

# Hapus baris yang memiliki NaN di kolom tertentu (misalnya "Nama")
df = df.dropna(subset=['Nama Produk'])

# Reset index setelah drop
df = df.reset_index(drop=True)
```

```python
df['Rating Produk'] = df['Rating Produk'].fillna(0).astype(float)
```

```python
# menyimpan data bersih pada csv
df.to_csv('dataset_P0GC3.csv')
```
